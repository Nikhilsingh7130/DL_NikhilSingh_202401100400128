```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error as mse
import seaborn as sns
```

```python
def sigmoid(z):
    return 1 /(1+ np.exp(-z))
```

```python
def derivative_sigmoid(z):
    s=sigmoid(z)
    return s*(1-s)
```

```python
X = [1, 2, 3, 4, 5]
y = []

for i in X:
    y.append(i * 2)
```

```python
X = np.array(X)
y = np.array(y)
```

```python
hd_w = np.random.rand(1)[0] 
hd_b = np.random.rand(1)[0]
op_w = np.random.rand(1)[0]
op_b = np.random.rand(1)[0]
```

```python
losses=[]
epoch=50
learning_rate=0.05
```

```python
for i in range(epoch):
    
    z1 = X * hd_w + hd_b
    a1 = sigmoid(z1)

    z2 = op_w * a1 + op_b
    y_pred = sigmoid(z2)

    
    loss = np.mean((y - y_pred) ** 2)
    print(f"Epoch {i+1}, Loss: {loss}")
    losses.append(loss)

    
    error = y_pred - y

    err_o = error * derivative_sigmoid(z2)

    hid_err = err_o * op_w
    err_hid = hid_err * derivative_sigmoid(z1)

    
    op_w = op_w - learning_rate * np.mean(err_o * a1)
    op_b = op_b - learning_rate * np.mean(err_o)

    hd_w = hd_w - learning_rate * np.mean(err_hid * X)
    hd_b = hd_b - learning_rate * np.mean(err_hid)
```
